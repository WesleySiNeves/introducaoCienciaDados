## Passo 1) Recuperar e pré processar os dados

### 1)  recuperar

```python
import pandas as pd
```

```python
arquivo ="/content/VendasSorvetes.txt"

dados =  pd.read_csv(arquivo)
dados.tail()
```

###  2) Pre-processar

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = dados["Temperatura"].values.reshape(-1,1)
y =  dados["Venda"].values
```

## Passo 2) Criar o modelo de M.L

```python
modelo = LinearRegression()
```

## Passo 3) Treinar o Modelo

```python
modelo.fit(X,y)
```

## Passo 4) Predição

```python
temperaturas = np.array([80,85]).reshape(-1,1)
```

```python
predicao = modelo.predict(temperaturas)
predicao
```

```python
plt.scatter(X,y, color ='blue', label='Dados')
plt.plot(X,modelo.predict(X), color ='red',label ='Linha da Regressão')
plt.xlabel("Temperatura")
plt.ylabel("Venda")
plt.title("Predição de Vendas por Regressão Linear")
plt.legend()
plt.show()
```

## Passo 6 ) Avaliacao

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_pred = modelo.predict(X)

MAE = mean_absolute_error(y,y_pred)

print(f"MEA: {MAE:.2f}")
```
